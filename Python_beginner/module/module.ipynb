```python
# 중복이 없는 1~45 사이의 random number 출력 
# 결과는 list 자료형으로 출력 
#방법 1 
import random 
lotto = []
while True:
    temp = random.randint(1,45)
    if temp not in lotto:  #중복을 막기 위해 
        lotto.append(temp)
    if len(lotto)==6:
        break

print(lotto)
```

```python
#방법 2 
l = []
while True: 
    l.append(random.randint(1,45))
    l = set(l) #set는 중복값을 가질 수 없다. 
    l = list(l)
    if len(l) == 6:
        break

print(l)
```
